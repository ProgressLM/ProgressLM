```python
import json

input_file = "/home/vcj9002/jianshu/chengxuan/ProgressLM/data/train/sft/sft_raw.jsonl"
output_file = "/home/vcj9002/jianshu/chengxuan/ProgressLM/data/train/sft/sft_raw_step.jsonl"

with open(input_file, 'r') as f:
    data = [json.loads(line) for line in f]

print(f"筛选前总数: {len(data)}")

filtered_data = [item for item in data if 3 <= int(item['total_steps']) <= 7]

print(f"筛选后总数: {len(filtered_data)}")
print(f"筛选掉: {len(data) - len(filtered_data)}")

with open(output_file, 'w') as f:
    for item in filtered_data:
        f.write(json.dumps(item, ensure_ascii=False) + '\n')

print(f"已保存到: {output_file}")
```
